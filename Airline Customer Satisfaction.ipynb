```python
Data3=Airline[['Gender','Customer Type','Type of Travel','Class','Satisfaction']]
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
```

```python
Data3.loc[:,:]=Data3.loc[:,:].apply(label_encoder.fit_transform)
```

```python
Data3.head()
```

```python
Airline[list(Data3.columns)]=Data3
Airline=Airline.apply(pd.to_numeric,errors='coerce')
Airline.head()
```

```python
corr=pd.DataFrame(Airline.corr()['Satisfaction'].drop('Satisfaction',axis=0))
corr.plot(kind='barh',grid=True,figsize=(20,20))
```

```python
features=[]
for i in corr.index:
    if corr.loc[i].values<-0.4:
        features.append(i)
    elif corr.loc[i].values>0.2:
        features.append(i)
```

```python
print(features)
```

```python
X=Airline[features]
y=Airline['Satisfaction']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
print("Training Accuracy",lr.score(X_train,y_train))
print("Testing Accuracy",lr.score(X_test,y_test))
```

```python
print('Logistic Regression Accuracy Score:',accuracy_score(pred,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
dt.pred=dt.predict(X_test)
```

```python
print("Training Accuracy",dt.score(X_train,y_train))
print("Testing Accuracy",dt.score(X_test,y_test))
```

```python
print('Decision Tree Accuracy Score:',accuracy_score(dt.pred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
rf.pred=rf.predict(X_test)
```

```python
print("Training Accuracy",rf.score(X_train,y_train))
print("Testing Accuracy",rf.score(X_test,y_test))
```

```python
print('Random Forest Accuracy Score:',accuracy_score(rf.pred,y_test))
```

```python
import xgboost 

from xgboost import XGBClassifier
xgb=XGBClassifier()
```

```python
xgb.fit(X_train,y_train)
```

```python
xgb.pred=xgb.predict(X_test)
```
